```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import psycopg2
import sqlite3
```

```python
#read the 2017-2018 data and put it into a df
last_year = "PGATOUR_data2017.csv"
lastseason_df = pd.read_csv(last_year)
lastseason_df.set_index("Player", inplace = True)
lastseason_df.dropna()
```

```python
#read the 2018-2019 data and put it into a df
this_year = "This_Season_Clean_data.csv"
thisseason_df = pd.read_csv(this_year)
thisseason_df.set_index("Player", inplace = True)
thisseason_df.head()
```

```python
#create db connection
# conn = psycopg2.connect(
#     host = "localhost", 
#     database="ETL-Golf", 
#     user="postgres", 
#     password="delasalle")

engine = create_engine('postgresql://postgres:delasalle@localhost/ETL-Golf')

```

```python
#load dataframes to the tables in PGAdmin
thisseason_df.to_sql(name='thisseason_df1', con=engine, if_exists='append', index=True)
```
