```python
import pandas as pd
```

```python
#read the csv and create a dataframe
this_year = "PGATOUR_data2019.csv"
thisyear_df = pd.read_csv(this_year)
thisyear_df.head(5)
```

```python
thisyear_df.columns
```

```python
#rename the columns we wand to use: AVG_Driving_DISTANCE, FAIRWAY_HIT_%, AVG_CARRY_DISTANCE, AVG_SCORE, WORLD_RANK 
thisyear_df = thisyear_df.rename(columns={'Driving Distance - (AVG.)':'AVG_Driving_Distance', 'Hit Fairway Percentage - (FAIRWAYS HIT)':'FAIRWAY_HIT_%', 'Carry Distance - (AVG.)':'AVG_CARRY_DISTANCE', 'Stroke Differential Field Average - (FIELD AVERAGE SCORE)':'AVG_SCORE', 'Official World Golf Ranking - (AVG POINTS)':'WORLD_RANK'})
```

```python
#create a df with only the coulmns in want in the final data set 
thisseason_df = thisyear_df.loc[:,['Player Name','AVG_Driving_Distance', 'FAIRWAY_HIT_%', 'AVG_CARRY_DISTANCE', 'AVG_SCORE','WORLD_RANK']]
thisseason_df
```

```python
#save df to csv in repo
thisseason_df.to_csv("This_Season_Clean_data.csv")
```
