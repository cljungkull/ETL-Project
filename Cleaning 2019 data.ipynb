```python
import pandas as pd
```

```python
#read the csv and create a dataframe
this_year = "PGATOUR_data2019.csv"
thisyear_df = pd.read_csv(this_year)
thisyear_df.head(5)
```

```python
thisyear_df.columns
```

```python
#rename the columns we wand to use: AVG_Driving_DISTANCE, FAIRWAY_HIT_%, AVG_CARRY_DISTANCE, AVG_SCORE, POINTS_BEHIND_LEAD, NUMBER_OF_TOP_Tens 
thisyear_df = thisyear_df.rename(columns={'Driving Distance - (AVG.)':'AVG_Driving_Distance', 'Player Name':'Player', 'Top 10 Final Round Performance - (EVENTS)':'NUMBER_OF_TOP_Tens', 'Carry Distance - (AVG.)':'AVG_CARRY_DISTANCE', 'Stroke Differential Field Average - (FIELD AVERAGE SCORE)':'AVG_SCORE', 'FedExCup Standings - (POINTS BEHIND LEAD)':'POINTS_BEHIND_LEAD'})
```

```python
#create a df with only the coulmns in want in the final data set 
thisseason_df = thisyear_df.loc[:,['Player','AVG_Driving_Distance', 'NUMBER_OF_TOP_Tens', 'AVG_CARRY_DISTANCE', 'AVG_SCORE','POINTS_BEHIND_LEAD', 'Season']]
thisseason_df
```

```python
#save df to csv in repo
thisseason_df.to_csv("This_Season_Clean_data.csv")
```
